```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# data шинч╜о
data = np.array([3, 12, 5, 18, 45])
bars = np.array(['A', 'B', 'C', 'D', 'E'])
y_pos = np.arange(len(bars))

plt.yticks([3,5,12,18,45]) # y ш╗╕чЪДхИ╗х║жхА╝
sns.barplot(y_pos, data, color=(0.2, 0.8, 0.6, 0.6))#color чВ║ RGB хПК Alpha шк┐цХ┤

plt.legend(['b'])  # х╖жф╕КхЬЦф╛Лц╖╗хКа

# Custom Axis title щЬАшжБшк┐цХ┤ matplotlib шинч╜о x , yх║зциЩ
plt.xlabel('title of the xlabel', fontweight='bold', color = 'orange', fontsize='17', horizontalalignment='center');
plt.ylabel('title of the ylabel ',fontweight='bold', color = 'r', fontsize='10', horizontalalignment='center')
```

```python
#шинхоЪх║ХхЬЦшГМцЩпцигх╝П
sns.set_style('darkgrid') # шинхоЪхЬЦх╜вцигх╝П
```
